```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Unemployment in India.csv")

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns=df.columns.str.strip()

df["Date"]=pd.to_datetime(
df["Date"],
dayfirst=True
)

df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
df["Estimated Unemployment Rate (%)"],
kde=True
)

plt.show()
```

```python
plt.figure(figsize=(15,6))

sns.barplot(
x="Region",
y="Estimated Unemployment Rate (%)",
data=df
)

plt.xticks(rotation=90)

plt.show()
```

```python
top=df.groupby(
"Region"
)[
"Estimated Unemployment Rate (%)"
].mean()

top=top.sort_values(
ascending=False
).head(10)

top.plot(
kind="bar"
)

plt.show()
```

```python
numeric=df.select_dtypes(
include="number"
)

sns.heatmap(
numeric.corr(),
annot=True,
cmap="coolwarm"
)

plt.show()
```
